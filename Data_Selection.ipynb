# Task #02 作業 01【實作題】資料選取

```python
#!pip install pandas
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print(df)
```

- 1. filtered by first column > 20?

```python
mask = df.iloc[:,0]>20
df[mask]
```

- 2. filtered by first column + second column > 50

```python
mask = df.iloc[:,0]+df.iloc[:,1] > 50
df[mask]
```

- 3. filtered by first column < 30 or second column > 30

```python
mask = (df.iloc[:,0]<30) | (df.iloc[:,1]>30)
df[mask]
```

- 4. filtered by total sum of row > 100

```python
mask = df.sum(axis=1) > 100
df[mask]
```

#練習：請利用 Pandas 計算給定的資料中不同 weekday 的 visitor 總和為何？

```python
import pandas as pd
import numpy as np

d = [
    {'city': 'Austin', 'visitor': 139, 'weekday': 'Sun'},
    {'city': 'Dallas', 'visitor': 237, 'weekday': 'Sun'},
    {'city': 'Austin', 'visitor': 326, 'weekday': 'Mon'},
    {'city': 'Dallas', 'visitor': 456, 'weekday': 'Mon'}
]

data = pd.DataFrame(d)
grouped = data.groupby('weekday')
grouped.agg({'visitor':'sum'})
```
